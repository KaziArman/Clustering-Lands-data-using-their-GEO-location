```python
#Importing the necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import glob
import plotly.express as px
from sklearn.cluster import KMeans
```

```python
mylist = [f for f in glob.glob("*.xlsx")]
mylist[0]
```

```python
#Data Processing
county_name = mylist[0]
geo_data = pd.read_excel(county_name) #read data

geo_data['combine'] = geo_data[['lat','lon']].values.tolist() #combine location
fin_list = geo_data['combine'].tolist() #list data
#fin_list

#geo_data.head(5)
geo_data.shape
```

```python
len(geo_data)/60
```

```python
#Clustering
number_of_clusters = 37 #input expected cluster number
km = KMeans(n_clusters=number_of_clusters) #define cluster numbers
km.fit(fin_list) #clustering
y_km = km.fit_predict(fin_list) #create cluster ID

#Reorder the cluster number
unique_order = pd.unique(y_km)
right_order = np.argsort(unique_order)
cluster = [ right_order[i] for i in y_km]
cluster = [x+1 for x in cluster]

#geo_data['cluster2'] = y_km
geo_data['cluster'] = cluster #add cluster id to dataset
```

```python
#export dataset
geo_data.drop('combine', axis=1, inplace=True)
geo_data = geo_data.sort_values(['cluster', '«APN»'], ascending=[True, True]) #sort data by cluster and then by APN

geo_data.to_excel(county_name, index=False)
```
