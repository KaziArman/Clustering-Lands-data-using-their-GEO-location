```python
#Importing the necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Data Processing
county_name = "Spartanburg, SC"
geo_data = pd.read_excel(county_name & ".xlsx") #read data
geo_data['combine'] = geo_data[['lat','lon']].values.tolist() #combine location
fin_list = geo_data['combine'].tolist() #list data
#fin_list

#geo_data.head(5)
geo_data.shape
```

```python
len(geo_data)/20
```

```python
#Finding optimum cluster value

#import libraries
from sklearn.cluster import KMeans
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import MinMaxScaler

#Calculate and plot elbow
Xl = geo_data['lat'].tolist()
Yl = geo_data['lon'].tolist()
Xa = np.array(Xl)
X = Xa.reshape(-1,1)
Ya = np.array(Yl)
Y = Ya.reshape(-1,1)

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = .3, random_state = 15)


# within clusters sum of squares 
wcss = []
K = [i for i in range(25,300,20)] #input range and leap
for k in K:
    kms = KMeans(n_clusters = k)
    kms = kms.fit(Y)
    wcss.append(kms.inertia_)
    
# Plotting the results onto a line graph, 
# `allowing us to observe 'The elbow'
plt.plot(K, wcss,'bx-')
plt.title('The elbow method')
plt.xlabel('No. of clusters')
plt.ylabel('WCSS') # Within cluster sum of squares
```

```python
#Clustering
number_of_clusters = 107 #input expected cluster number
km = KMeans(n_clusters=number_of_clusters) #define cluster numbers
km.fit(fin_list) #clustering
y_km = km.fit_predict(fin_list) #create cluster ID

#Reorder the cluster number
unique_order = pd.unique(y_km)
right_order = np.argsort(unique_order)
cluster = [ right_order[i] for i in y_km]
cluster = [x+1 for x in cluster]

#geo_data['cluster2'] = y_km
geo_data['cluster'] = cluster #add cluster id to dataset
```

```python
#Plotting clusters
data_with_clusters = geo_data.copy()
data_with_clusters['clusters'] = y_km 
plt.scatter(data_with_clusters['lon'],data_with_clusters['lat'],c=data_with_clusters['clusters'],cmap='rainbow')
```

```python
#export dataset
geo_data = geo_data.sort_values(['cluster', '«APN»'], ascending=[True, True]) #sort data by cluster and then by APN
geo_data.to_excel('Spartanburg, SC.xlsx', index=False)
```
